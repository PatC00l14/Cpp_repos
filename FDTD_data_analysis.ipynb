```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animate

dat = pd.read_csv('ftdt_save.csv').to_numpy()

def convert(data):
    norm = 1
    return(data[::2,:]/norm,data[1::2,:-1]/norm)

E_y, H_z = convert(dat)

xspace = np.arange(0, len(E_y[0,:]),1)

axs = plt.figure(figsize = (15, 8 )).add_subplot(projection = '3d')


for i in range(500,1000, 500):
    axs.plot(xspace[:-1], H_z[i,:],zs = 0, color = 'blue')
    axs.plot(xspace, 0, zs = E_y[i,:], color = 'red')

axs.plot(1,1,color = 'red', label = 'E_y')
axs.plot(1,1,color = 'blue', label = 'H_z')
axs.set_xlabel('X-steps') ; axs.set_ylabel('Amplitude');axs.set_zlabel('Amplitude')
axs.axvline(35, linestyle = 'dotted') ; axs.legend()
axs.set_title('1D FDTD with exponential pulse')


```
