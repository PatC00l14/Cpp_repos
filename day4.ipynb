```python
import numpy as np
import matplotlib.pyplot as plt


day4 = open('xmas_txt.txt')
xmas_txt = day4.read()
day4.close()

fig, axs = plt.subplots(1,1)

l_xmas = 140

xmas_rows = []



method =1

if method ==1:
    xmas_rows.append(xmas_txt[0:l_xmas])
    for i in range(1, l_xmas):
        xmas_rows.append(xmas_txt[i*l_xmas + i:(i+1)*l_xmas+i])

elif method ==2:
    xmas_txt = xmas_txt.strip('/n')
    for i in range(l_xmas):
        xmas_rows.append(
            xmas_txt[i*l_xmas:(i+1)*l_xmas]
        )

for xr in xmas_rows:
    if len(xr)!= 140:
        print(f'XXXXXXXXXXX wrong {len(xr)}')

xcoords = []
count = 0
for r in range(l_xmas):
    for c in range(l_xmas):
        count +=1
        if xmas_rows[r][c] != 'S' and xmas_rows[r][c] != 'X' and xmas_rows[r][c] != 'M' and xmas_rows[r][c] != 'A':
            print('fucked up paddy',xmas_rows[r][c])
        if xmas_rows[r][c] == 'X' or xmas_rows[r][c] == 'S': #identified a possible xmas
            if c<l_xmas-3:
                xstr = xmas_rows[r][c:c+4]
                #Check to the RIGHT or LEFT
                if xstr == 'XMAS' or xstr == 'SAMX':
                    xcoords.append([r,c,1])
            
            if r <l_xmas-3:
                #Check UP or DOWN
                xstr = xmas_rows[r][c] + xmas_rows[r+1][c] + xmas_rows[r+2][c] + xmas_rows[r+3][c]
                if xstr == 'XMAS' or xstr =='SAMX':
                    xcoords.append([r,c,2])
            
            if c<l_xmas-3 and r>2:
                #Check UP diagonal - so 45/ 225 degress
                xstr = xmas_rows[r][c] + xmas_rows[r-1][c+1] + xmas_rows[r-2][c+2] + xmas_rows[r-3][c+3]
                if xstr == 'XMAS' or xstr =='SAMX':
                    xcoords.append([r,c,3])
            
            if c<l_xmas-3 and r<l_xmas-3:
                #Check down diagonals - so 135 / 315 degrees
                xstr = xmas_rows[r][c] + xmas_rows[r+1][c+1] + xmas_rows[r+2][c+2] + xmas_rows[r+3][c+3]
                if xstr == 'XMAS' or xstr =='SAMX':
                    xcoords.append([r,c,4])


for x in xcoords:
    if True:
        if x[2]==1:
            axs.plot(x[0], x[1], 'ro', markersize = 0.9, color = 'b') 
        if x[2]==2:
            axs.plot(x[0], x[1], 'ro', markersize = 0.9, color = 'r') 
        if x[2]==3:
            axs.plot(x[0], x[1], 'ro', markersize = 0.9, color = 'g') 
        if x[2]==4:
            axs.plot(x[0], x[1], 'ro', markersize = 0.9, color = 'black') 
axs.set_title(f'Xmas - {len(xcoords)}')

print(f'count = {count}')

```
